# Histogram Transformation
    Histogram of an image transform, the x-axis shows the gray level intensities and the y-axis shows the gray level intensities and y-axis shows the frequency of these intensities.
    mostly used in thresholding

```python
import cv2
from matplotlib import pyplot as plt

# reads an input image
img = cv2.imread('../Images/blurred21.jpg',0)
plt.imshow(img)
# find frequency of pixels in range 0-255
histr = cv2.calcHist([img],[0],None,[256],[0,256])


plt.subplot(2, 2, 1)
plt.title('image')
plt.imshow(img,cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 2)
plt.title('histogram')
plt.plot(histr)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python

# alternative way to find histogram of an image
plt.hist(img.ravel(),256,[0,256])
plt.show()
```

```python
#equalized the image
equalized=cv2.equalizeHist(img)

fig=plt.figure(dpi=300)
fig.add_subplot(1,3,1)
plt.imshow(img,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,3,2)
plt.imshow(equalized,cmap='gray')
plt.axis("off")
plt.title("Equalized")
```
