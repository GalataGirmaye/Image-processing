```python
import cv2
from matplotlib import pyplot as plt

# reads an input image
img = cv2.imread('../Images/lenaLC.jpg',0)
plt.imshow(img)
# find frequency of pixels in range 0-255
histr = cv2.calcHist([img],[0],None,[256],[0,256])


plt.subplot(2, 2, 1)
plt.title('image')
plt.imshow(img,cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 2)
plt.title('histogram')
plt.plot(histr)
plt.xticks([])
plt.yticks([])
plt.show()
```
