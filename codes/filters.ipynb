```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread("../Images/lenna.jpeg",0)
img = cv2.resize(img, (225, 225),interpolation=cv2.INTER_CUBIC)
```

```python
def showpic(img1,img2,img2_name):
    fig=plt.figure(dpi=300)

    fig.add_subplot(1,3,1)
    plt.imshow(img1,cmap='gray')
    plt.axis("off")
    plt.title("Original")

    fig.add_subplot(1,3,2)
    plt.imshow(img2,cmap='gray')
    plt.axis("off")
    plt.title(img2_name)
```

## Image Filtering

Filtering techniques are used to enhance and modify digital images. Also, images filters are use to blurring and noise reduction , sharpening and edge detection. Image filters are mainly use for suppress high (smoothing techniques) and low frequencies(image enhancement, edge detection).
Image Filtering techniques are categorized as follows:
### 1. Spatial Domain
#### 1. Linear Filter

#### Mean Filter
The Averaging Filter technique takes the average of all the pixels under the kernel area and replaces the central element. This eliminates some of the noise in the image and smooths the edges of the image. The blur function from the Open-CV library can be used to apply a mean filter to an image.

```python
#(5,5) indicated the kernel size
averaged = cv2.blur(img,(5,5))
fig=plt.figure(dpi=300)

fig.add_subplot(1,3,1)
plt.imshow(img,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,3,2)
plt.imshow(averaged,cmap='gray')
plt.axis("off")
plt.title("Averaged")
```

#### Gaussian Filter
It is also low pass filter. The Gaussian Filter is similar to the mean filter however it involves a weighted average of the surrounding pixels and has a parameter sigma. The kernel represents a discrete approximation of a Gaussian distribution. We can use gaussian filter in denoising (gaussian noise can denoised by gaussian filter). Gaussian filter smoothen the picture.

```python
gaussian= cv2.GaussianBlur(img, (21, 21),cv2.BORDER_DEFAULT)

fig=plt.figure(dpi=300)

fig.add_subplot(1,3,1)
plt.imshow(img,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,3,2)
plt.imshow(gaussian,cmap='gray')
plt.axis("off")
plt.title("Gaussian")
```

#### Laplasian Filter
The Laplacian filter detects sudden intensity transitions in the image and highlights the edges.There are two kinds of laplacian kernels. The one including diagonals the other not including diagonals. Since the Laplacian filter detects the edges of an image it can be used along with a Gaussian filter in order to first remove speckle noise and then to highlight the edges of an image. This method is referred to as the Lapalcian of Gaussian filtering.

```python
#prepare the 5x5 shaped filter
# kernel = np.array([[0,-1,0], 
#                    [-1,4,-1],
#                    [0,-1,-1]])
#kernel = kernel/sum(kernel)


laplacian = cv2.Laplacian(img, cv2.CV_64F)
showpic(img,laplacian,'Laplacian')
```
